```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_stocks = pd.read_csv('./stockPricePaths.csv',header=None)
df_options = pd.read_csv('./optionPricePaths.csv', header=None)
df_errors = pd.read_csv('./hedgingErrorPaths.csv', header=None)
```

```python
df_errors.describe()
```

```python
df_errors.drop(df_errors.columns[-1], axis=1, inplace=True)
data = df_errors.to_numpy().flatten()
data
```

```python
plt.figure(figsize=(10, 6))
for index, row in df_stocks.iterrows():
    plt.plot(row.index, row.values)

# Optional: Display a legend if you want to identify each line
plt.legend()

plt.xlabel('Time steps')
plt.ylabel('Stock Prices')
plt.title('Stock Price Paths')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for index, row in df_options.iterrows():
    plt.plot(row.index, row.values)

# Optional: Display a legend if you want to identify each line
plt.legend()

plt.xlabel('Time steps')
plt.ylabel('Option Prices')
plt.title('Option Price Paths')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data, bins=50, density=True, color='skyblue', edgecolor='black')
plt.title('Hedging Error Distribution')
plt.show()
```
